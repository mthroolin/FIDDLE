## Test 1: discretize = FALSE, imputation_method =  FFILL

```python
import os
import sys
import io
import subprocess
from pathlib import Path

cur_dir = os.getcwd()

# Set console to UTF-8
os.system('chcp 65001')

# Get current directory, which works on both Unix and Windows
base_dir = os.path.abspath(os.path.join(cur_dir, '../../'))

# Set the PYTHONPATH
python_path = os.environ.get('PYTHONPATH', '')
os.environ['PYTHONPATH'] = f"{python_path}{separator}{base_dir}" if python_path else base_dir
```

```python
# Create a directory using system-agnostic methods
output_dir = Path("output-FFILL")
output_dir.mkdir(parents=True, exist_ok=True)

command = [
    'C:\\Users\\mthro\\anaconda3\\python.exe',
    '-m', 'FIDDLE.run',
    '--data_fname=input\\data.csv',
    '--population_fname=input\\pop.csv',
    '--config_fname=input\\config-1-parallel.yaml',
    '--output_dir=output-FFILL',
    '--T=4', '--dt=1.0',
    '--theta_1=0.001', '--theta_2=0.001', '--theta_freq=1',
    '--stats_functions=min', '--stats_functions=max', '--stats_functions=mean', '--impute_method=ffill'
]

result = subprocess.run(command, capture_output=True, text=True)

# Print stdout and stderr
print("Standard Output:\n", result.stdout)
```

```python
import numpy as np
import pandas as pd
import json
import sparse

S_ffill = sparse.load_npz('output-FFILL/S_all.npz')
S_names_ffill = json.load(open('output-FFILL/S_all.feature_names.json', 'r'))
S_index_ffill = pd.read_csv('output-FFILL/S.ID.csv').set_index(['ID'])
df_S_ffill = pd.DataFrame(S_ffill.todense(), columns=S_names_ffill, index=S_index_ffill.index)

X_ffill = sparse.load_npz('output-FFILL/X_all.npz')
X_names_ffill = json.load(open('output-FFILL/X_all.feature_names.json', 'r'))
X_index_ffill = pd.read_csv('output-FFILL/X.ID,t_range.csv').set_index(['ID', 't_range'])
df_X_ffill = pd.DataFrame(X_ffill.todense().reshape(-1, X_ffill.shape[-1]), columns=X_names_ffill, index=X_index_ffill.index)

display(df_S_ffill)
display(df_X_ffill)
```
